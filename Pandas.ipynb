```python
type(data)
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object)
```

```python
df
```

```python
df['a'][1]
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64'})
```

```python
df
```

```python
df['a'][1]
```
